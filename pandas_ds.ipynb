## 5. 모든 행과 특정 열 데이터을 함께 추출하기

### 5.1 슬라이싱 사용

```python
# loc로 year와 pop 열 데이터 추출하여 변수 sub_df에 저장하기
# 이때 콜론으로 행을 지정하여 year와 pop 열의 모든 행 데이터을 가져오기
sub_df = df.loc[:,['year','pop']]
sub_df
```

```python
# 데이터 3, 5번째와 마지막 열 데이터 추출하기
# iloc로 콜론을 사용하여 모든 행 데이터 가져오기
sub_df = df.iloc[: ,[2,4,-1]]
sub_df
```

### 5.2 iloc 속성과 range함수 사용하여 데이터 추출
* range()에 시작값과 끝값을 입력하면 자동으로 시작값 이상, 끝값 미만인 값을 생성
* 간격을 지정하여 생성하는 숫자를 조절할 수 있다.
* range()함수는 리스트가 아닌 정해진 범위의 수를 반환하는 제너레이터.
* 제너레이터는 그때마다 값을 생성하며 한 번 사용한 값은 메모리에서 사라짐

```python
# 1. range(n)의 반환값 0이상 n 미만의 정수를 반환하는 제너레이터 생성
# 2. 제너레이터를 리스트로 변환
# 3. iloc 속성에 사용
small_range = list(range(5))
sub_df = df.iloc[:, small_range]
sub_df
```

```python
# 모든 행의 3,4,5 번 열을 추출하기 : range함수
small_range = list(range(3,6))
sub_df = df.iloc[:, small_range]
sub_df
```

### 5.3 iloc 속성과 슬라이싱 구문을 사용하여 데이터 추출
* range함수에 전달한 인자를 : 기호로 구분하여 대괄호에 작성

```python
# 슬라이싱 구문 으로 추출
sub_df = df.iloc[:, :3]
sub_df
```

```python
# 모든 행의 3 이상과 6 미만의 범위에 해당하는 열 데이터 추출
sub_df = df.iloc[:, 3:6]
sub_df
```

### 5.4 여러 행과 열 지정하여 데이터 추출하기
* 열 데이터 추출은 모든 행을 대상 
    * 모두 loc와iloc 의 대괄호에서 쉼표 왼쪽에 : 기호 사용
* 특정 행과 열에 있는 데이터 추출
    * 모두 loc와iloc 의 대괄호에서 쉼표 왼쪽에 값 변경

```python
# iloc 사용 : 첫 번째, 4번째, 6번째 열의 첫 번째, 100번째, 1000번째 행 데이터 추출하기
sub_df = df.iloc[[0,99,999],[0,3,5]]
sub_df
```

```python
# loc 사용 : 첫 번째, 4번째, 6번째 열의 첫 번째, 100번째, 1000번째 행 데이터 추출하기
sub_df = df.loc[[0,99,999],['country','lifeExp','gdpPercap']]
sub_df
```

```python
# iloc & loc 사용 : 10번째부터 13번째 행의 모든 열 데이터 추출하기
#<주의> 
# loc는 콜론 오른쪽 번호를 결과에 포함
# iloc는 콜론 오른쪽 번호를 결과에에서 제외
sub1_df = df.loc[10:13, :]
print(sub1_df)
print()
sub2_df = df.iloc[10:13, :]
print(sub2_df)
```

# 6. 데이터 집계
* 추출한 데이터을 대상으로 기초 통계 계산
* 그룹화한 데이터의 평균
* 2 개 이상의 열 그룹화 하기

```python
df
```

### <1. 열 그룹화 하기>
* groupby메서드 사용

```python
# 연도별 기대 수명 계산
# 1. 데이터를 연도별 그룹화
grouped_year_df = df.groupby('year') #DataFrameGroupBy
type(grouped_year_df)

# 2. lifeExp 열을 선택
grouped_year_df_lifeExp = grouped_year_df['lifeExp'] # SeriesGroupBy
type(grouped_year_df_lifeExp)

# 3. 평균 구하기
mean_lifeExp_by_year = grouped_year_df_lifeExp.mean()
mean_lifeExp_by_year
```

```python
# 한 줄 명령 구문
mean_lifeExp_by_year = df.groupby('year')['lifeExp'].mean()
mean_lifeExp_by_year
```

### <2. 2개 이상 열 그룹화 하기>
* 여러 열에 같은 연산을 적용
* 다층 계층 구조로 구성


```python
# 1. 연도와 대륙별로 데이처 그룹화
# 2. 연도에 따른 대륙별 기대수명과 GDP 평균 계산

multi_group_var = df.groupby(['year','continent'])[['lifeExp','gdpPercap']].mean()
multi_group_var
```

```python
# 처음 10개 행 보기
multi_group_var.head()
```
